```python
import glob
import os
import pandas as pd
```

```python
first_half = glob.glob("Feb2021-May2021/*.csv")
first_df = pd.concat((pd.read_csv(f) for f in first_half))
first_df
```

```python
first_df.dropna()#.count()
```

```python
first_df.to_csv('First Half.csv')
```

```python
second_half = glob.glob("Jun2021-Dec2021/*.csv")
second_df = pd.concat((pd.read_csv(f) for f in second_half))
second_df
```

```python
second_df.dropna()#.count()
```

```python
second_df.to_csv('Second Half.csv')
```

```python
third_part = glob.glob("Jan2022-April2022/*.csv")
third_df = pd.concat((pd.read_csv(f) for f in third_part))
third_df
```

```python
third_df.dropna()#.count()
```

```python
third_df.to_csv('Third Part.csv')
```

```python
all_df = pd.concat([first_df, second_df, third_df]).reset_index(drop=True)
all_df
```

```python
all_df.to_csv('Full Data.csv')
```
